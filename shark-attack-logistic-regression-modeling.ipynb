```python
#see if it is worthwhile to use regex to separate into categories
import regex as re
def find_phrase(p):
    return(lambda x: re.findall(r'{}'.format(p), x))


def get_phrase_count(phrase, data, col):
    data_new = data.dropna() #added dropna to ignore null entries
    phrase_results = data_new[col].apply(find_phrase(phrase)) 
    phraselist = list(filter(None, phrase_results))
    phraseCount = len(phraselist)
    return phraseCount

def get_df_record(phrase, data, col):
    data_new = data.dropna() #added dropna to ignore null entries
    phrase_results = data_new[col].dropna().apply(find_phrase(r'{}'.format(phrase))) 
    phrase_bool = pd.Series([bool(x) for x in phrase_results], index = data_new.index)
    df_recs = data_new[phrase_bool]
    return df_recs
```

Train and Test Logistic Regression Model

```python
from sklearn.model_selection import train_test_split
train_data , test_data = train_test_split(sharkdf, test_size=0.2, random_state=38)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import logit 
```

```python
train_data
```

Use of all variables prompted 'Maximum number of iterations has been exceeded' error. Geographical data had to be excluded in order to avoid the error. The data set was already shown to have several issues with structure and consistency, making it difficult to use in a comprehensive way.

```python
formula = ('Fatal ~ Type + Sex')

model = logit(formula = formula, data = train_data).fit()
```

```python
model.summary()
```

```python
#logistic regression interpretation tutorial
#https://onezero.blog/modelling-binary-logistic-regression-using-python-research-oriented-modelling-and-interpretation/

import numpy as np
print(np.exp(model.params))
```

This shows how much the lack of data skews the results of the logistic regression. If we were to interpret this without context, the invalid category indicates one unit increase in the invalid category of attack type increases the odds of a fatal attack by 358 times.

This is a frustrating example of where the sample size is not the issue for modeling correctly, but the lack of complete and consistent data.
Near infinite categories of geographical data instead of consolidation into succinct groups resulted in the removal from the model. Hence defeating the original objective of determining fatal attacks by location, area, or country, at the minimum, along with type and sex.
